# Set

Creating Set

```python
s = {1,2,3,4,5}

print(s)

s2 = set([1,2,3,4,5])

print(s2)

s3 = set("abc")

print(s3)
```

Looping through Set

```python
s = {1,2,3,4,5}

for element in s:
  print(element)
```

Adding element in Set

```python
s = {1,2,3,4,5}

s.add(6)

print(s)
```

Adding multiple elemnts from another set to a set

```python
s1 = {1,2,3}

s2 = {4,5,6}

s1.update(s2)

print(s1)
```

Removing element from Set

```python
s = {1,2,3,4,5}

s.remove(4)

print(s)

s.discard(5)

print(s)

s.pop()

print(s)
```

# Dictionary

Creating dictionary

```python
d = {"node1":10,"node2":20,"node3":30}

print(d)

d2 = dict([("node1",10),("node2",20),("node3",30)])

print(d2)
```

Accessing Dictionary

```python
d = {"node1":10,"node2":20,"node3":30}

print(d["node1"])
print(d.get("node2"))
print(d.keys())
print(d.values())
print(d.items())
```

Changing dictionary

```python
d = {"node1":10,"node2":20,"node3":30}

d["node1"]=15

print(d)

d.update({"node2":25})

print(d)
```

Adding Items in dictionary

```python
d = {"node1":10,"node2":20,"node3":30}

d.update({"node4":40})

print(d)
```

Deleting Items from dictionary

```python
d = {"node1":10,"node2":20,"node3":30}

d.pop("node1")

print(d)

d.popitem()

print(d)

del d["node2"]

print(d)
```

Looping through dictionary

```python
d = {"node1":10,"node2":20,"node3":30}

for key in d:
  print(key)
  print(d[key])

for key,value in d.items():
  print(key)
  print(value)

for key in d.keys():
  print(key)

for value in d.values():
  print(value)
```

# **Numpy**

## Creating 1D array

```python
import numpy as np

temperatures = np.array([10,20,30,40,50])

print(temperatures)

print(type(temperatures))

print(temperatures.shape)

print(temperatures.dtype)

print(temperatures.ndim)

print(temperatures.size)
```

## Creating a 2D array

```python
temp_and_humidity = np.array([[10,20,30],[40,50,60]])

print(temp_and_humidity)

print(type(temp_and_humidity))

print(temp_and_humidity.shape)

print(temp_and_humidity.dtype)

print(temp_and_humidity.ndim)

print(temp_and_humidity.size)
```

## Accessing elements from 1D array

```python
temperatures = np.array([10,20,30,40,50])

print(temperatures[0])
print(temperatures[1])

print(temperatures[-1])
print(temperatures[-2])
```

## Accessing elements from 2D array

```python
temp_and_humidity = np.array([[10,20,30],[40,50,60]])

print(temp_and_humidity[0,2])

print(temp_and_humidity[1,2])

print(temp_and_humidity[1,-1])
```

## Slicing 1D array

```python
temperatures = np.array([10,20,30,40,50])

print(temperatures[1:4])

print(temperatures[1:4:2])

print(temperatures[1:])

print(temperatures[:3])

print(temperatures[-3:-1])

print(temperatures[-3:-1:2])

print(temperatures[::2])
```

## Slicing 2D array

```python
temp_and_humidity = np.array([[10,20,30],[40,50,60]])

print(temp_and_humidity[0,1:3])

print(temp_and_humidity[1,1:])

print(temp_and_humidity[0:2,2])

print(temp_and_humidity[:,1])

print(temp_and_humidity[0:2,1:3])

print(temp_and_humidity[:,::2])

print(temp_and_humidity[::2,:])
```

## Reshaping Arrays

```python
temperatures = np.array([10,20,30,40,50,60])

print(temperatures.reshape(2,3))

print(temperatures.reshape(3,2))
```

## Arithmetic functions on Numpy arrays

```python
temperatures_1 = np.array([10,20,30,40,50])

temperatures_2 = np.array([5,10,15,20,25])

print(np.add(temperatures_1,temperatures_2))

print(np.subtract(temperatures_1,temperatures_2))

print(np.multiply(temperatures_1,temperatures_2))

print(np.divide(temperatures_1,temperatures_2))

print(np.mod(temperatures_1,temperatures_2))
```

## Summation over different axis in Numpy

```python
temperatures_1 = np.array([10,20,30,40,50])

temperatures_2 = np.array([5,10,15,20,25])

print(np.sum([temperatures_1,temperatures_2]))

print(np.sum([temperatures_1,temperatures_2],axis=0))

print(np.sum([temperatures_1,temperatures_2],axis=1))
```

## Concatenation of Two 1D arrays

```python
temperatures_1 = np.array([10,20,30,40,50])

temperatures_2 = np.array([5,10,15,20,25])

print(np.concatenate((temperatures_1,temperatures_2)))  # same shape

print(np.stack((temperatures_1,temperatures_2),axis=1)) # have one more dimension than input dimension

print(np.hstack((temperatures_1,temperatures_2)))

print(np.vstack((temperatures_1,temperatures_2)))

print(np.dstack((temperatures_1,temperatures_2)))
```

## Concatenation of Two 2D arrays

```python
temp_and_humidity_1 = np.array([[10,20,30],[40,50,60]])

temp_and_humidity_2 = np.array([[5,10,15],[20,25,30]])

print(np.concatenate((temp_and_humidity_1,temp_and_humidity_2),axis=0))

print(np.concatenate((temp_and_humidity_1,temp_and_humidity_2),axis=1))

print(np.stack((temp_and_humidity_1,temp_and_humidity_2),axis=1))

print(np.hstack((temp_and_humidity_1,temp_and_humidity_2)))

print(np.vstack((temp_and_humidity_1,temp_and_humidity_2)))

print(np.dstack((temp_and_humidity_1,temp_and_humidity_2)))
```

# **Pandas**

## Creating Dataframe

```python
import pandas as pd

details = {
    "node_1":[10,20,30],
    "node_2":[40,50,60],
    "node_3":[70,80,90]
}

df = pd.DataFrame(details)

print(df)

print(type(df))

print(df.shape)

print(df.size)

print(df.ndim)

print(df.columns)

print(df.index)

print(df.dtypes)

print(df.info())

print(df.head(2))

print(df.tail(2))

print(df.describe())

print(df.T)
```

## Setting Index Name

```python
df = pd.DataFrame(details,index=["temp","humidity","pressure"])

print(df)

print(df.index)
```

## Locating row in dataframe

```python
print(df.loc["temp"])

print(df.loc["temp","node_1"])

print(df.loc[["temp","pressure"]])

print(df.loc[["temp","pressure"],["node_1","node_2"]])
```
